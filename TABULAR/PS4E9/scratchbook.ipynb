```python
text1 = "2.7L V6 24V pdi dohc Twin Turbo"
text2 = "395.0HP 5.7L 8 cylinder Engine gasoline/Mild electric Hybrid"
# relevant subtext to extract number of cylinders is 6 cylinders and V8
cyl_pattern = r"\d+\s*cylinder|V\s*\d+"
fuel_type_pattern = r'(gasoline|diesel|flex|electric)|(pdi dohc|gdi dohc)'
cyl_match = re.search(cyl_pattern, text1)
cyl_match_all = re.findall(cyl_pattern, text1)
fuel_match1 = re.search(fuel_type_pattern, text1)
fuel_match2 = re.search(fuel_type_pattern, text2)
print(cyl_match.group(0))
print(fuel_match1.group(0))
print(fuel_match2.group(0))
print(cyl_match_all)
```

```python
transmission_list = [x.lower() for x in transmission_list]
auto_pattern = r"automatic|cvt|a/t|\sat|transmission overdrive switch"
manual_pattern = r"manual|m/t|\smt"
single_speed_pattern = r"single-speed"
speed_pattern = r"(\d+)[\s-]speed"

speed_match = re.search(speed_pattern, "6 speed Mt")
if speed_match:
    print(speed_match.group(1))

for x in transmission_list:
    auto_match = re.search(auto_pattern, x)
    manual_match = re.search(manual_pattern, x)
    single_speed_match = re.search(single_speed_pattern, x)
    if auto_match:
        transmission_type = "automatic"
    elif manual_match:
        transmission_type = "manual"
    elif single_speed_match:
        transmission_type = "single_speed"
    elif x == "F":
        transmission_type = "automatic"
    else:
        transmission_type = "unknown"
    print(transmission_type)
```
